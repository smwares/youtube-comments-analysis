```python
import os
import numpy as np
import pandas as pd
from dotenv import load_dotenv
from googleapiclient.discovery import build
from pprint import pprint
```

```python
# get the API key from the .env file
load_dotenv()
api_key = os.getenv('API_KEY')
```

```python
# insert user ID to extract user info from (this would be the text that comes after 'https://www.youtube.com/')
userId = '@spider-manunknown9193'

# initiate the YouTube client
youtube = build(serviceName='youtube', version='v3', developerKey=api_key)

part_list = ['brandingSettings', 'contentDetails', 'contentOwnerDetails','id', 'localizations', 'snippet', 'statistics', 'status', 'topicDetails']

# create a request to get the parent comments, then save the response
for part in part_list:
    request = youtube.channels().list(part=part, forHandle=userId, maxResults=50)
    response = request.execute()
    print('\n' + '-'*50, '\n', part, '\n' + '-'*50 + '\n')
    pprint(response)
```
